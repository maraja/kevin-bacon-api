```python
def degreesOfSeparation(graph, start, end, maxDepth=-1):
    visited = []
    queue = []
    visited.append(start)
    queue.append(start)
    queue.append(None)
    level = 0

    while queue:
        s = queue.pop(0) 
        if s == None:
            level += 1
        elif s == end:
            print(f'Found {end}! Degrees of separation: {level}')
            return
        else:
            graph[s].append(None)
            for neighbour in graph[s]:
                if neighbour not in visited:
                    visited.append(neighbour)
                    queue.append(neighbour)
        if level == maxDepth:
            print("Max Depth reached, could not find degree of separation.")
            queue = []
            return
```

## Helpers

```python
dedupe = lambda t: list(set(t))
flatten = lambda t: [item for sublist in t for item in sublist]
```

```python
class ActorsService():
    def __init__(self):
        self.credits_df = pd.read_csv("./credits.csv")
        
    def getActorsById(self, id: int):
        actors = []
        for cast_member in eval(credits_df[credits_df.id == int(id)].iloc[0]["cast"]):
            actors.append(cast_member["name"])
        return actors
```
